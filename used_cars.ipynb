```python
import pandas as pd

df = pd.read_csv("used_cars.csv")

# Basic inspection
print(df.shape)
print(df.info())
print(df.describe(include='all'))
print(df.head())
```

```python
from sklearn.metrics import mean_absolute_error

# Mean prediction
mean_price = df['selling_price'].mean()
predictions = [mean_price] * len(df)

# Calculate MAE
mae = mean_absolute_error(df['selling_price'], predictions)

print("Baseline MAE:", mae)
```

```python
# 1. Drop rows where target is missing
df = df.dropna(subset=['selling_price'])

# 2. Standardize brand column
df['brand'] = df['brand'].str.strip().str.lower()

# 3. Clean mileage (extract numeric values)
df['mileage'] = df['mileage'].str.extract('(\d+\.?\d*)').astype(float)

# 4. Convert selling_price to numeric (if needed)
df['selling_price'] = pd.to_numeric(df['selling_price'], errors='coerce')

# 5. Impute missing values (numerical columns)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# 6. Final check
print(df.isnull().sum())
```
